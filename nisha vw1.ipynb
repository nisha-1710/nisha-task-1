```python
# Import libraries
import pandas as pd
```

```python
# Load dataset
df = pd.read_csv("Medical_Dataset.csv")
df
```

```python
# 1. Explore Dataset
print("First 5 Records")
print(df.head())


print("\nDataset Shape:")
print(df.shape)


print("\nColumn Names:")
print(df.columns)


print("\nData Types:")
print(df.dtypes)
```

```python
# 2. Missing Values
print("Missing Values:")
print(df.isnull().sum())
```

```python
# Fill missing values

df['Medicine_1'].fillna(df['Medicine_1'].mode()[0], inplace=True)

df['Dosage_1'].fillna(df['Dosage_1'].mode()[0], inplace=True)

df['Duration_1'].fillna(df['Duration_1'].mode()[0], inplace=True)

df['Instructions_1'].fillna(df['Instructions_1'].mode()[0], inplace=True)
```

```python
print("Missing Values:")
print(df.isnull().sum())
```

```python
# 3. Duplicate Records

duplicates = df.duplicated().sum()
print("\nDuplicate Records:", duplicates)
```
